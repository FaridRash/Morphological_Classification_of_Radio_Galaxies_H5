```python
import tensorflow as tf

MODEL_PATH = "/content/Morphological_Classification_of_Radio_Galaxies_H5/Morphological_Classification_The_final_code.h5"  # Corrected path

model = tf.keras.models.load_model(MODEL_PATH)

print("Input shape:", model.input_shape)
print("Output shape:", model.output_shape)

# Optional but useful:
model.summary()
```

```python
import numpy as np
from PIL import Image

IMG_H, IMG_W = 60, 60

def preprocess_pil_to_model_input(img: Image.Image) -> np.ndarray:
    """
    Returns: x with shape (1, 60, 60, 1), dtype float32.
    Normalization: [0,1] based on /255.0
    """
    img = img.convert("L")               # force grayscale
    img = img.resize((IMG_W, IMG_H))     # resize to 60x60 (W,H)
    x = np.array(img).astype(np.float32) # (60,60)
    x = x / 255.0                        # normalize
    x = x[..., None]                     # (60,60,1)
    x = np.expand_dims(x, axis=0)        # (1,60,60,1)
    return x
```

```python
dummy = Image.fromarray(np.random.randint(0, 256, (60, 60), dtype=np.uint8), mode="L")
x = preprocess_pil_to_model_input(dummy)

print("x.shape:", x.shape, "x.dtype:", x.dtype, "min/max:", float(x.min()), float(x.max()))

y = model.predict(x)
print("y.shape:", y.shape)
print("raw y:", y)
print("sum probs (if softmax):", float(np.sum(y[0])))
print("predicted index:", int(np.argmax(y[0])))
```

```python
import numpy as np

IMG_H, IMG_W = 60, 60

def preprocess_step_only(img):
    img_l = img.convert("L")                # force grayscale
    img_r = img_l.resize((IMG_W, IMG_H))    # resize to 60×60
    x = np.array(img_r).astype("float32")   # (60,60)
    x = x / 255.0                           # normalize [0,1]
    x = x[..., None]                        # (60,60,1)
    x = x[None, ...]                        # (1,60,60,1)
    return x

x = preprocess_step_only(img)

print("Final tensor shape:", x.shape)
print("dtype:", x.dtype)
print("min / max:", float(x.min()), float(x.max()))
```

```python
import matplotlib.pyplot as plt

# x was created in Part 3C with shape (1, 60, 60, 1)
cropped_img = x[0, :, :, 0]  # remove batch & channel dims → (60,60)

plt.figure(figsize=(3,3))
plt.imshow(cropped_img, cmap="gray")
plt.title("Model Input (60×60, Grayscale)")
plt.axis("off")
plt.show()

print("Shape shown:", cropped_img.shape)
print("Min / Max:", float(cropped_img.min()), float(cropped_img.max()))
```

```python
import numpy as np

y = model.predict(x)

print("Raw output vector:", y[0])
print("Shape:", y.shape)

pred_idx = int(np.argmax(y[0]))
print("Predicted class index:", pred_idx)
print("Sum of outputs:", float(np.sum(y[0])))
```
