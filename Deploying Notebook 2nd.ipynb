```python
import tensorflow as tf

MODEL_PATH = "/content/Morphological_Classification_of_Radio_Galaxies_H5/Morphological_Classification_The_final_code.h5"  # Corrected path

model = tf.keras.models.load_model(MODEL_PATH)

print("Input shape:", model.input_shape)
print("Output shape:", model.output_shape)

# Optional but useful:
model.summary()
```

```python
import numpy as np
from PIL import Image

IMG_H, IMG_W = 60, 60

def preprocess_pil_to_model_input(img: Image.Image) -> np.ndarray:
    """
    Returns: x with shape (1, 60, 60, 1), dtype float32.
    Normalization: [0,1] based on /255.0
    """
    img = img.convert("L")               # force grayscale
    img = img.resize((IMG_W, IMG_H))     # resize to 60x60 (W,H)
    x = np.array(img).astype(np.float32) # (60,60)
    x = x / 255.0                        # normalize
    x = x[..., None]                     # (60,60,1)
    x = np.expand_dims(x, axis=0)        # (1,60,60,1)
    return x
```

```python
dummy = Image.fromarray(np.random.randint(0, 256, (60, 60), dtype=np.uint8), mode="L")
x = preprocess_pil_to_model_input(dummy)

print("x.shape:", x.shape, "x.dtype:", x.dtype, "min/max:", float(x.min()), float(x.max()))

y = model.predict(x)
print("y.shape:", y.shape)
print("raw y:", y)
print("sum probs (if softmax):", float(np.sum(y[0])))
print("predicted index:", int(np.argmax(y[0])))
```

```python
from fastapi import FastAPI, File, UploadFile
import io
import numpy as np
from PIL import Image
import tensorflow as tf

app = FastAPI()

IMG_H, IMG_W = 60, 60

def preprocess_for_model(file_bytes: bytes) -> np.ndarray:
    img = Image.open(io.BytesIO(file_bytes)).convert("L")
    img = img.resize((IMG_W, IMG_H))
    x = np.array(img).astype("float32") / 255.0
    x = x[..., None]      # (60,60,1)
    x = x[None, ...]      # (1,60,60,1)
    return x

@app.get("/health")
def health():
    return {"status": "ok"}

@app.post("/predict")
async def predict(file: UploadFile = File(...)):
    b = await file.read()
    x = preprocess_for_model(b)

    y = model.predict(x)[0]
    probs = tf.nn.softmax(y).numpy().tolist()
    pred = int(np.argmax(probs))

    return {
        "predicted_index": pred,
        "probabilities": probs
    }
```

```python
import threading
import uvicorn

def run():
    uvicorn.run(app, host="0.0.0.0", port=8000, log_level="info")

thread = threading.Thread(target=run, daemon=True)
thread.start()

print("Server started on port 8000")
```

```python
%%writefile app/main.py
import io
import os
import numpy as np
from fastapi import FastAPI, File, UploadFile
from PIL import Image
import tensorflow as tf

# ---- Config ----
MODEL_PATH = os.environ.get(
    "MODEL_PATH",
    "/content/Morphological_Classification_of_Radio_Galaxies_H5/Morphological_Classification_The_final_code.h5"
)
IMG_H, IMG_W = 60, 60

# Optional: put the true class names if you want
CLASS_NAMES = os.environ.get("CLASS_NAMES", "FR0,FRI,FRII").split(",")

app = FastAPI()

# ---- Load model once at startup ----
model = tf.keras.models.load_model(MODEL_PATH)

def preprocess_for_model(file_bytes: bytes) -> np.ndarray:
    img = Image.open(io.BytesIO(file_bytes)).convert("L")
    img = img.resize((IMG_W, IMG_H))
    x = np.array(img).astype("float32") / 255.0
    x = x[..., None]      # (60,60,1)
    x = x[None, ...]      # (1,60,60,1)
    return x

@app.get("/health")
def health():
    return {"status": "ok"}

@app.post("/predict")
async def predict(file: UploadFile = File(...)):
    b = await file.read()
    x = preprocess_for_model(b)

    y = model.predict(x)[0]
    probs = tf.nn.softmax(y).numpy()
    pred = int(np.argmax(probs))

    label = CLASS_NAMES[pred] if pred < len(CLASS_NAMES) else str(pred)

    return {
        "predicted_index": pred,
        "predicted_label": label,
        "probabilities": probs.tolist()
    }
```
